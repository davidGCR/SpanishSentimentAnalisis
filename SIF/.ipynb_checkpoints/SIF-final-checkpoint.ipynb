```python
#retornar indice de palabra en vocab
def lookupIDX(words,w):
    w = w.lower()
    if len(w) > 1 and w[0] == '#':
        w = w.replace("#","")
    if w in words:
        #retornar indice de palabra en vocab
        return words[w]
    elif 'UUUNKKK' in words:
        return words['UUUNKKK']
    else:
        return len(words) - 1

class tree(object):

    def __init__(self, phrase, words):
        self.phrase = phrase
        self.embeddings = []
        self.representation = None

    #reemplaza oracion con array de indices
    def populate_embeddings(self, words):
        #concertir frase a minusculas
        phrase = self.phrase.lower()
        
        #dividir en palabras
        arr = phrase.split()
#         print(phrase,len(arr))
        for token in arr:
            self.embeddings.append(lookupIDX(words,token))

    def unpopulate_embeddings(self):
        self.embeddings = []
```

```python
def getSentimentDataset(f,words):
    data = open(f,'r')
    lines = data.readlines()
    examples = []
    for idx,i in enumerate(lines):
        i=i.strip()
        if(len(i) > 0):
            i=i.split('\t')
            if len(i) == 2:
                e = (tree(i[0], words), i[1])
                examples.append(e)
            else:
                print('o noooo: ',i, idx)
    return examples
```

```python
def getWeight(words, word2weight):
    weight4ind = {}
    for word, ind in words.items():
        if word in word2weight:
            weight4ind[ind] = word2weight[word]
        else:
            weight4ind[ind] = 1.0
    return weight4ind
```

```python
#Crea dictionary: {'word': freq inverse SIF,}
def getWordWeight(weightfile, a=1e-3):
    if a <=0: # when the parameter makes no sense, use unweighted
        a = 1.0

    word2weight = {}
    with open(weightfile) as f:
        lines = f.readlines()
    N = 0
    for idx,i in enumerate(lines):
        i=i.strip()
        if(len(i) > 0):
            i=i.split()
            if(len(i) == 2):
                word2weight[i[0]] = float(i[1])
                N += float(i[1])
            else:
                print('joderrrr ',i, idx)
    for key, value in word2weight.items():
        word2weight[key] = a / (a + value/N)
    return word2weight
```

```python
############### Parameters #####################
pc = []
npc = 0
layersize = 100 ##2400
nonlinearity = lasagne.nonlinearities.linear
memsize = 150
LW = 1e-05 
LC = 1e-06 
updatewords = False ##Actualizar SIF embeddings
clip = None
learner = lasagne.updates.adagrad 
eta = 0.05 
task = "sentiment"
```

```python
class lasagne_average_layer(lasagne.layers.MergeLayer):
    
    def __init__(self, incoming, **kwargs):
        super(lasagne_average_layer, self).__init__(incoming, **kwargs)
    
    def get_output_for(self, inputs, **kwargs):
        emb = inputs[0]
        mask = inputs[1]
        emb = (emb * mask[:, :, None]).sum(axis=1)
        emb = emb / mask.sum(axis=1)[:, None]
        return emb
    
    def get_output_shape_for(self, input_shape):
        return (input_shape[0][0],input_shape[0][2])
    
def getRegTerm(We, initial_We):
    l2 = 0.5*LC*sum(lasagne.regularization.l2(x) for x in network_params)
    if updatewords:
        return l2 + 0.3*LW*lasagne.regularization.l2(We-initial_We)
    else:
        return l2

def getTrainableParams():
    if updatewords:
        return all_params
    else:
        return network_params
```

```python
#We_initial = We
initial_We = theano.shared(np.asarray(We, dtype = config.floatX))
We = theano.shared(np.asarray(We, dtype = config.floatX))
if npc > 0:
    pc = theano.shared(np.asarray(pc, dtype = config.floatX))

g1batchindices = T.imatrix()
g1mask = T.matrix()
scores = T.matrix()

l_in = lasagne.layers.InputLayer((None, None))
l_mask = lasagne.layers.InputLayer(shape=(None, None))
l_emb = lasagne.layers.EmbeddingLayer(l_in, input_size=We.get_value().shape[0], output_size=We.get_value().shape[1], W=We)
# l_emb = lasagne.layers.EmbeddingLayer(l_in, input_size=We.shape[0], output_size=We.shape[1], W=We)
l_average = lasagne_average_layer([l_emb, l_mask])

l_out = lasagne.layers.DenseLayer(l_average,layersize, nonlinearity=nonlinearity)
embg = lasagne.layers.get_output(l_out, {l_in:g1batchindices, l_mask:g1mask})

############# Model Variables ####################
network_params= None
all_params= None
trainable= None
feedforward_function= None
scoring_function= None
cost_function= None
train_function= None

if npc <= 0:
    print ("#pc <=0, do not remove pc")
elif npc == 1:
    print ("#pc == 1")
    proj =  embg.dot(pc.transpose())
    embg = embg - theano.tensor.outer(proj, pc)
else:
    print ("#pc > 1")
    proj =  embg.dot(pc.transpose())
    embg = embg - theano.tensor.dot(proj, pc)
    
############# Model ####################
l_in2 = lasagne.layers.InputLayer((None, layersize))
# l_sigmoid = lasagne.layers.DenseLayer(l_in2, memsize, nonlinearity=lasagne.nonlinearities.sigmoid)
# l_softmax = lasagne.layers.DenseLayer(l_sigmoid, 2, nonlinearity=T.nnet.softmax)
l_sigmoid = lasagne.layers.DenseLayer(l_in2, memsize, nonlinearity=lasagne.nonlinearities.softmax)
l_softmax = lasagne.layers.DenseLayer(l_sigmoid, 4, nonlinearity=T.nnet.softmax)



X = lasagne.layers.get_output(l_softmax, {l_in2:embg})
cost = T.nnet.categorical_crossentropy(X,scores)

prediction = T.argmax(X, axis=1)

network_params = lasagne.layers.get_all_params(l_out, trainable=True) + lasagne.layers.get_all_params(l_softmax, trainable=True)
network_params.pop(0) # do not include the word embedding as network parameters
all_params = lasagne.layers.get_all_params(l_out, trainable=True) + lasagne.layers.get_all_params(l_softmax, trainable=True)

reg = getRegTerm(We, initial_We)
trainable = getTrainableParams()
cost = T.mean(cost) + reg

##functions
feedforward_function = theano.function([g1batchindices,g1mask], embg)
scoring_function = theano.function([g1batchindices, g1mask],prediction)
cost_function = theano.function([scores, g1batchindices, g1mask], cost)

grads = theano.gradient.grad(cost, trainable)
if clip:
    grads = [lasagne.updates.norm_constraint(grad, clip, range(grad.ndim)) for grad in grads]
updates = learner(grads, trainable, eta)
train_function = theano.function([scores, g1batchindices, g1mask], cost, updates=updates)
```

```python
################ Train util fucntions #####################

## Padding sentences and buil mask 
def prepare_data(list_of_seqs):
    lengths = [len(s) for s in list_of_seqs]# lista de longitudes de oraciones
    n_samples = len(list_of_seqs)#numero de batch
    maxlen = np.max(lengths)# oracion de maxima longitud
    x = np.zeros((n_samples, maxlen)).astype('int32') #matrix embedding? padding
    x_mask = np.zeros((n_samples, maxlen)).astype('float32')

    for idx, s in enumerate(list_of_seqs):
        x[idx, :lengths[idx]] = s #padding de indices [32, 31, 4, 1, 5, 0, 0, 0, ...]
        x_mask[idx, :lengths[idx]] = 1.#mascara [1.0, 1.0, 1.0, 1.0, 0 , 0, 0....]
    x_mask = np.asarray(x_mask, dtype='float32')
    return x, x_mask

    
##build sentence to list of indeces   
def getSeq(p1,words):
    p1 = p1.split()
    X1 = []
    for i in p1:
        X1.append(lookupIDX(words,i))
    return X1


#g1mask = seq2weight(g1x, g1mask, weight4ind)
#seq = [34 2 12 3 5 0 0 0 ...]: 
#construye matrix de embbddings de una oracion pero en ves de indices de vocabulario tiene los SIF de las palabras
#para una oracion: [0.2212,1.0, 0.31343, ... ]
def seq2weight(seq, mask, weight4ind):
    weight = np.zeros(seq.shape).astype('float32')
    for i in range(seq.shape[0]):
        for j in range(seq.shape[1]):
            if mask[i,j] > 0 and seq[i,j] >= 0:
                #weight4ind: {'word': freq inverse SIF,}
                weight[i,j] = weight4ind[seq[i,j]]#get SIF of word
    weight = np.asarray(weight, dtype='float32')#(numero_ejemplos, long_ maxima de oracion)
    return weight


def getAccSentiment(words,f):
    f = open(f,'r')
    lines = f.readlines()
    preds = []##predictions
    golds = [] ##real labels
    seq1 = []# store sequences (list of indeces)
    ct = 0
    for i in lines:
        i = i.split("\t")
        p1 = i[0]; ##sentence
        score = i[1]##real label
        X1 = getSeq(p1,words) ##convert sentence to list of indeces
        seq1.append(X1)##store sequences
        ct += 1
        if ct % 100 == 0:
            x1,m1 = prepare_data(seq1)
            
            m1 = seq2weight(x1, m1, weight4ind)
            
            scores = scoring_function(x1,m1)
#             print('predicitions: ',m1)
            
            scores = np.squeeze(scores)
            preds.extend(scores.tolist())##add predictions
            seq1 = []
           
        golds.append(score)
    if len(seq1) > 0: ##the rest of examples
        x1,m1 = prepare_data(seq1)
        
        m1 = seq2weight(x1, m1, weight4ind)
        scores = scoring_function(x1,m1)
        scores = np.squeeze(scores)
        preds.extend(scores.tolist())
    return accSentiment(preds,golds)

def supervised_evaluate(words,file):
    #if params.task == "sentiment":
    s = getAccSentiment(words,file)
    return s

def get_minibatches_idx(n, minibatch_size, shuffle=False):
    idx_list = np.arange(n, dtype="int32")

    if shuffle:
        np.random.shuffle(idx_list)

    minibatches = []
    minibatch_start = 0
    for i in range(n // minibatch_size):
        minibatches.append(idx_list[minibatch_start:
        minibatch_start + minibatch_size])
        minibatch_start += minibatch_size

    if (minibatch_start != n):
        minibatches.append(idx_list[minibatch_start:])

    return zip(range(len(minibatches)), minibatches)

def getDataSentiment(batch):
    g1 = []
    for i in batch:
        g1.append(i[0].embeddings)

    g1x, g1mask = prepare_data(g1)

    scores = []
    for i in batch:
        temp = np.zeros(4)
        label = i[1].strip()
        if label == "0":
            temp[0]=1
        elif label == "1":
            temp[1]=1
        elif label == "2":
            temp[2]=1
        elif label == "3":
            temp[3]=1
        else: 
            print("label no valido")
#         print("SCORES: ",label)
        scores.append(temp)
    scores = np.matrix(scores)+0.000001
    scores = np.asarray(scores,dtype='float32')
    return (scores,g1x,g1mask)

# def getDataSentiment(batch):
#     g1 = []
#     for i in batch:
#         g1.append(i[0].embeddings)

#     g1x, g1mask = prepare_data(g1)

#     scores = []
#     for i in batch:
#         temp = np.zeros(2)
# #         label = i[1].strip()
# #         if label == "0":
# #             temp[0]=1
# #         if label == "1":
# #             temp[1]=1
# #         temp = np.zeros(4)
#         label = i[1].strip()
#         if label == "0":
#             temp[0]=1
#         elif label == "1":
#             temp[1]=1
#         else: 
#             print("label no valido")
# #         if label == "2":
# #             temp[2]=1
# #         if label == "3":
# #             temp[3]=1
# #         print("SCORES: ",label)
#         scores.append(temp)
#     scores = np.matrix(scores)+0.000001
#     scores = np.asarray(scores,dtype='float32')
#     return (scores,g1x,g1mask)

# def accSentiment(preds,scores):
#     golds = []
#     for n,i in enumerate(scores):
#         p = -1
#         i=i.strip()
# #         print('label read: ',i)
#         if i == "0":
#             p = 0
#         elif i == "1":
#             p = 1
# #         elif i == "2":
# #             p = 2
# #         elif i == "3":
# #             p = 3
#         else:
#             raise ValueError('Something wrong with data...')
#         golds.append(p)
    
#     return accuracy_score(golds,preds)

def accSentiment(preds,scores):
    golds = []
    for n,i in enumerate(scores):
        p = -1
        i=i.strip()
        if i == "0":
            p = 0
        elif i == "1":
            p = 1
        elif i == "2":
            p = 2
        elif i == "3":
            p = 3
        else:
            raise ValueError('Something wrong with data...')
        golds.append(p)
    
    return accuracy_score(golds,preds)
    
```

```python
################# Train ##########################
def train_util(train_data, dev, test, train, words, epochs):
    print("utility function for training the model")
    start_time = time()
    try:
        for eidx in range(epochs):
            kf = get_minibatches_idx(len(train_data), batchsize, shuffle=True)
#             kf = get_minibatches_idx(1000, batchsize, shuffle=True)
            uidx = 0
            for _, train_index in kf:
                uidx += 1
                batch = [train_data[t] for t in train_index]
                # load the word ids
                for i in batch:
                    i[0].populate_embeddings(words)
                # load the data
                (scores,g1x,g1mask) = getDataSentiment(batch)

#                 print('g1x: ',g1x)
#                 print('glmax: ',g1mask)
#                 print('scores: ',scores)
                
                g1mask = seq2weight(g1x, g1mask, weight4ind)
                cost = train_function(scores, g1x, g1mask)
                if np.isnan(cost) or np.isinf(cost):
                    print ('NaN detected')
                # undo batch to save RAM
                for i in batch:
                    i[0].representation = None
                    i[0].unpopulate_embeddings()
            ts = supervised_evaluate(words,test)
            rs = supervised_evaluate(words,train)
#             print ("evaluation: (dev,test, train)",ds,ts,rs)
            print ("evaluation: (test, train)",ts,rs)
            print ('Epoch ', (eidx+1), 'Cost ', cost)
            sys.stdout.flush()
    except KeyboardInterrupt:
        print ("Training interupted")
    end_time = time()
    print ("total time:", (end_time - start_time))
```

```python
def make_prediction(words,f):
    f = open(f,'r')
    lines = f.readlines()
    preds = []##predictions
    golds = [] ##real labels
    seq1 = []# store sequences (list of indeces)
    ct = 0
    for i in lines:
        i = i.split("\t")
        p1 = i[0]; ##sentence
        score = i[1]##real label
        golds.append(int(score))
        
        #print(score)
        X1 = getSeq(p1,words) ##convert sentence to list of indeces
        seq1.append(X1)##store sequences
        ct += 1
        if ct % 100 == 0:
            x1,m1 = prepare_data(seq1)
            
            m1 = seq2weight(x1, m1, weight4ind)
            
            scores = scoring_function(x1,m1)
#             print('predicitions: ',m1)
            
            scores = np.squeeze(scores)
            preds.extend(scores.tolist())##add predictions
            seq1 = []
           
        
    if len(seq1) > 0: ##the rest of examples
        x1,m1 = prepare_data(seq1)
        
        m1 = seq2weight(x1, m1, weight4ind)
        scores = scoring_function(x1,m1)
        scores = np.squeeze(scores)
        preds.extend(scores.tolist())
    return golds,preds

def make_prediction_2(words,f):
    f = open(f,'r')
    lines = f.readlines()
    preds = []##predictions
#     golds = [] ##real labels
    seq1 = []# store sequences (list of indeces)
    ct = 0
    for i in lines:
        i = i.split("\t")
        p1 = i[0]; ##sentence
#         score = i[1]##real label
#         golds.append(int(score))
        
        #print(score)
        X1 = getSeq(p1,words) ##convert sentence to list of indeces
        seq1.append(X1)##store sequences
        ct += 1
        if ct % 100 == 0:
            x1,m1 = prepare_data(seq1)
            
            m1 = seq2weight(x1, m1, weight4ind)
            
            scores = scoring_function(x1,m1)
#             print('predicitions: ',m1)
            
            scores = np.squeeze(scores)
            preds.extend(scores.tolist())##add predictions
            seq1 = []
           
        
    if len(seq1) > 0: ##the rest of examples
        x1,m1 = prepare_data(seq1)
        
        m1 = seq2weight(x1, m1, weight4ind)
        scores = scoring_function(x1,m1)
        scores = np.squeeze(scores)
        preds.extend(scores.tolist())
    return preds
```
