### Xml to Csv parsing

```python
import xml.etree.ElementTree as ET
import csv

def xml2csv(xml_file,csv_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()

    data = open(csv_file, 'w')

    csvwriter = csv.writer(data)
    head = []

    count = 0
    for member in root.findall('tweet'):
        tweet = []
        address_list = []
        
        id_t=int(member.find('tweetid').text)
        print(id_t)
        tweet.append(id_t)

        content = member.find('content').text
        tweet.append(content)

        sentiment = member.find('sentiment')[0][0].text

        tweet.append(sentiment)

        csvwriter.writerow(tweet)
    data.close()
```
